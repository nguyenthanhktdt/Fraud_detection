```python
import pandas as pd
from imblearn.combine import SMOTEENN
from sklearn.model_selection import train_test_split

# Features and target
X = df[['ip_is_suspicious', 'device_is_suspicious', 'is_short_time']]
y = df['class']

# Apply SMOTEENN
smote_enn = SMOTEENN(random_state=42)
X_resampled, y_resampled = smote_enn.fit_resample(X, y)

# Convert back to DataFrame for easier handling
balanced_data = pd.DataFrame(X_resampled, columns=X.columns)
balanced_data['class'] = y_resampled

# Output the balanced class distribution
print("Balanced class distribution:\n", balanced_data['class'].value_counts())
```

```python
balanced_data.to_csv('/content/drive/MyDrive/Vistula_BDA/df_train_balanced.csv', index=False)
```
